## 2. Variable Importance

```python
import numpy as np
import matplotlib.pyplot as plt

betaList = np.array([[0.5], [-2], [2], [0.1], [4]])
matX = np.array([[1, 1, 2, 10, 0.1], [1, 0, 1, -8, -0.02], [1, 2, 0, 4, 0.04], [1, 2.5, -1, 9, 0.1]])
matY = np.array([[0.8], [0.6], [0.1], [0.2]])
cr = ['r', 'b', 'g' , 'm']

plt.figure(figsize = (10, 8))
for i in range(4):
    tmp = matX.copy()
    x_i = np.linspace(matX[:, i + 1].min(), matX[:, i + 1].max(), 20)
    y_p = []
    for j in x_i:
        tmp[:, i + 1] = j
        y_p.append(np.mean(1 / (1 + np.exp(-tmp@betaList))))
    plt.subplot(2, 2, i+1)
    plt.plot(x_i, y_p, cr[i])
    plt.xlabel('x_' + str(i+1))
    plt.ylabel('y_partial' + str(i+1))
plt.tight_layout()
plt.show()
```
