```python
import numpy as np

matX = np.array([[1,1,0], [2,0,1], [1,0,-0.5]], dtype = float)
matY = np.array([[1.5], [5], [1.5]], dtype = float)
betaList = np.array([[0], [0], [0]], dtype = float)
print("matX: ", matX)
print("Betas: ", betaList)
print("matY: ", matY)
val_a = 1
max_iterr = 3
matF = matX.shape[1]
activeMat = []
inactMat = list(np.arange(0, matF, 1)) 
for i in range(max_iterr + 1):
    print("\nIter # " + str(i+1) + " :")
    C_k = matX.T @ (matY - matX @ betaList)
    print("Ck: ", C_k)
    C_k[activeMat, :] = -np.inf 
    A_max = np.argmax(C_k)
    activeMat.append(A_max)
    print("Active Set: ", activeMat)
    for am in activeMat:
        if am in inactMat:
            inactMat.remove(am)
    print("Inactive Set: ", inactMat)
    alphaK = matX[:, activeMat].reshape(matF, len(activeMat))
    print("Ak: ", alphaK)
    print("Beta_Ak: ", betaList[activeMat])
    Y_hat_k = (alphaK @ betaList[activeMat]).reshape(matF, 1)
    print("Yhat_k: ", Y_hat_k)
    r_k = matY - Y_hat_k
    print("Rk: ", r_k)
    print("c: ", alphaK.T @ r_k)
    C_hat = np.max(np.abs(alphaK.T @ r_k))
    print("C_hat: ", C_hat)
    d_k = np.linalg.inv(alphaK.T @ alphaK) @ alphaK.T @ r_k
    print("Dk: ", d_k)
    Uk = alphaK @ d_k
    print("Uk: ", Uk)
    if i == 3: 
        alpha_k = val_a
    else:
        alphaList = []
        print("Calulcation of Ak")
        for j in inactMat:
            X_A_c = matX[:, j].reshape(matF, 1)
            Y_hat_c = (X_A_c @ betaList[j]).reshape(matF, 1)
            print("Y_hat_c" + str(j) + ": ", Y_hat_c)

            r_c = matY - Y_hat_c
            print("r_c_" + str(j) + ": ", r_c)
            c_j = X_A_c.T @ r_c
            print("c_" + str(j) + ": ", c_j)
            alphaList.append(float((C_hat - c_j) / (C_hat - X_A_c.T @ Uk)))
            alphaList.append(float((C_hat + c_j) / (C_hat + X_A_c.T @ Uk)))
        print("Alpha values: ", alphaList)
        alphaList = np.array(alphaList)
        alpha_k = np.min(alphaList[alphaList > 0])
    print("Ak: ", alpha_k)
    
    if len(activeMat) == 1:
        betaList[activeMat] = betaList[activeMat] + (alpha_k * float(d_k))
    else:
        betaList[activeMat] = betaList[activeMat] + (alpha_k * d_k)
    print("New beta val: ", betaList)
    
    Y_hat_k = Y_hat_k + (alpha_k * Uk)
    print("New Yhat_k val: ", Y_hat_k)
```
